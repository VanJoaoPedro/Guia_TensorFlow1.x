```
import input_data
import tensorflow.compat.v1 as tf
tf.disable_control_flow_v2()
tf.compat.v1.disable_eager_execution()
```

```
mnist = input_data.read_data_sets('mnist/', one_hot=True)
```

```
X_train = mnist.train.images
X_train.shape
```

```
y_train = mnist.train.labels
y_train.shape
```

```
X_test = mnist.test.images
X_test.shape
```

```
y_test = mnist.test.labels
y_test.shape
```

```
X_batch, y_batch = mnist.train.next_batch(128)
X_batch.shape
```

```
input_neurons = X_train.shape[1]
```

```
hidden_neurons1 = int((X_train.shape[1] + y_train.shape[1]) / 2)
hidden_neurons2 = hidden_neurons1
hidden_neurons3 = hidden_neurons1
output_neurons = y_train.shape[1]
```

```
weights = {'hidden_layer1': tf.Variable(tf.random_normal([input_neurons, hidden_neurons1])),
           'hidden_layer2': tf.Variable(tf.random_normal([hidden_neurons1, hidden_neurons2])),
           'hidden_layer3': tf.Variable(tf.random_normal([hidden_neurons2, hidden_neurons3])),
           'output_layer': tf.Variable(tf.random_normal([hidden_neurons3, output_neurons])),
           }
```

```
bias ={'hidden_layer1': tf.Variable(tf.random_normal([hidden_neurons1])),
       'hidden_layer2': tf.Variable(tf.random_normal([hidden_neurons2])),
       'hidden_layer3': tf.Variable(tf.random_normal([hidden_neurons3])),
       'output_layer': tf.Variable(tf.random_normal([output_neurons])),
        }
```

```
X_placeholder = tf.placeholder('float', [None, input_neurons])
y_placeholder = tf.placeholder('float', [None, output_neurons])
```

```
def model(x, weights, bias):
    hidden_layer1 = tf.nn.relu(tf.add(tf.matmul(x, weights['hidden_layer1']), bias['hidden_layer1'])) 
    hidden_layer2 = tf.nn.relu(tf.add(tf.matmul(hidden_layer1, weights['hidden_layer2']), bias['hidden_layer2'])) 
    hidden_layer3 = tf.nn.relu(tf.add(tf.matmul(hidden_layer2, weights['hidden_layer3']), bias['hidden_layer3'])) 
    output_layer =  tf.add(tf.matmul(hidden_layer3, weights['output_layer']), bias['output_layer']) 
    return output_layer
```

```
model = model(X_placeholder, weights, bias)
error = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=model, labels=y_placeholder))
optimizer = tf.train.AdadeltaOptimizer(learning_rate=0.0001).minimize(error)
```

```
prevision = tf.nn.softmax(model)
correct_prevision = tf.equal(tf.argmax(prevision, 1), tf.argmax(y_placeholder, 1))
hit_rate = tf.reduce_mean(tf.cast(correct_prevision, tf.float32))
```

```
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epochs in range(10000):
        X_batch, y_batch = mnist.train.next_batch(128)
        _, cost = sess.run([optimizer, error], feed_dict = {X_placeholder: X_batch, y_placeholder: y_batch})
        if epochs % 100 == 0:
            print('epochs : ' + str((epochs + 1)) + ' error : ' + str(cost))

    print(sess.run(hit_rate, feed_dict={X_placeholder: X_test, y_placeholder:y_test}))
```
