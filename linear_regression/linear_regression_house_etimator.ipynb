```
import pandas as pd
import tensorflow as tf
import tensorflow.compat.v1 as tf
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
tf.disable_v2_behavior()
```

```
base = pd.read_csv('/content/drive/My Drive/TensorFlow_1.x_guide/linear_regression/CSV/house_prices.csv')
base.head()
```

```
X = base.iloc[:, 5:6].values
y = base.iloc[:, 2:3].values
```

```
X
```

```
y
```

```
scaler_X = StandardScaler()
scaler_y = StandardScaler()
```

```
X = scaler_X.fit_transform(X)
y = scaler_y.fit_transform(y)
```

```
X
```

```
y
```

```
columns = [tf.feature_column.numeric_column('x', shape=1)]
columns
```

```
regressor = tf.estimator.LinearRegressor(feature_columns=columns)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```
X_train.shape
```

```
y_train.shape
```

```
X_test.shape
```

```
y_test.shape
```

```
function_train = tf.estimator.inputs.numpy_input_fn({'x': X_train}, y_train, batch_size=32, num_epochs= None, shuffle=True)
```

```
function_test = tf.estimator.inputs.numpy_input_fn({'x': X_test}, y_test, batch_size=32, num_epochs=1000, shuffle=False)
```

```
regressor.train(input_fn = function_train, steps = 10000)
```

```
metrics_train = regressor.evaluate(input_fn=function_train, steps=1000)
```

```
metric_test = regressor.evaluate(input_fn=function_test, steps=1000)
```

```
metrics_train
```

```
metric_test
```

```
new_home = np.array([[800], [900], [1000]])
```

```
new_home
```

```
new_home = scaler_X.transform(new_home)
```

```
function_prevision = tf.estimator.inputs.numpy_input_fn({'x': new_home}, shuffle=False)
```

```
prevision = regressor.predict(input_fn=function_prevision)
```
