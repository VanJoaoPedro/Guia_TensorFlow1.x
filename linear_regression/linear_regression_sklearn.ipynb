```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
X = np.array([[18], [23], [28], [33], [38], [43], [48], [58], [53], [63]])
```

```python
X
```

```python
y = np.array([[871], [1132], [1042], [1356], [1488], [1569], [1638], [1754], [1866], [1900]])
```

```python
y
```

```python
%matplotlib inline
plt.scatter(X, y)
```

```python
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
# b0 VALOR INICIAL DA LINHA
regressor.intercept_
```

```python
# b1 DECLIVE DA LINHA
regressor.coef_
```

```python
prevision1 = regressor.intercept_ + regressor.coef_ * 40
```

```python
prevision1
```

```python
predictors = regressor.predict(X)
predictors
```

```python
result = abs(y - predictors).mean()
result
```

```python
mae = mean_absolute_error(y, predictors)
mse = mean_squared_error(y, predictors)
mae
```

```python
mse
```

```python
plt.plot(X, y, 'o')
plt.plot(X, predictors, '*', color='red')
plt.title('Linear_regression_simple')
plt.xlabel('Age')
plt.ylabel('Price')
```
