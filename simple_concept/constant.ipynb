```
import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```
val1 = tf.constant(2)
val2 = tf.constant(3)
```

```
type(val1)
```

```
print(val2)
```

```
soma = val1 + val2
```

```
type(soma)
```

```
print(soma)
```

```
with tf.Session() as sess:
  s = sess.run(soma)
```

```
print(s)
```
